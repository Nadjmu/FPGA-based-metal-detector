```python
import numpy as np
import matplotlib.pyplot as plt

U = 15 #[V]
#I = 5 #[A]
f = 100 #[Hz]
r = 0.25 #[m]
a = 0.021 #[m]    #length of solenoid
#d = 0.5 #[mm]


#constants
pi = 3.14
mu = 4*pi*(10**-7)                              #N/(A^2)
rho = 1.72 * 10**-8 #1/sig                      #resistivity [Ome * m]

d = 1.32
d = d/1000 ##mm -> m

N = a/d
print("Number of turns:               ", N)
print(" ")

def calculateRea():
    A_sol = pi*r**2                            #cross section of the coil
    l = 2*pi*r                                 #average length of one coil
    L = np.power(N,2) * mu * A_sol/a           #inductance
    w = 2*pi*f
    Rea = w*L                                  #Reactance
    print("Reactance:                 ", Rea)
    print(" ")
    return Rea

def calculateRes():
    A_wire = pi * np.power(d/2, 2)             #cross section of wire total 
    l_tot = N*(2*pi*r)                         #total length of the wire
    Res = rho*l_tot/A_wire                     #resistivity
    print("Resistance:                ", Res)
    print(" ")
    return Res

def calculateZ(Rea, Res):
    Z = np.sqrt(np.power(Res,2) + np.power(Rea,2))
    print("Z:          ", Z)
    print(" ")
    return Z

def calculateI(Z):
    I = U/Z
    print("I:          ", I)
    print(" ")
    return I

def calculateB_inside(I):
    B = mu*I*N/a
    print("Magnetic field B inside the coil:      ", B)
    return B

def calculateB_outside(I, z): #calculate the magnetic field outside on z-axis
    B = mu * N * r **2* I /(2* np . power ( r **2+ z **2 , 3/2) )
    print("Magnetic field B outside the coil:     ", B)
    print(" ")
    return B

Rea = calculateRea()
Res = calculateRes()
Z = calculateZ(Rea,Res)
I = calculateI(Z)
B = calculateB_outside(I, 0.115)


########################## calculating the parameters of the Helmholtz-compensation solenoid 
c = np.power(4/5,3/2)
R = 0.03

def calculate_N(B,I):
    N = B/c * (R/(mu * I))
    print("number of turns needed:   ", N)
    print(" ")
    return N

N = calculate_N(B,I)

    
```
